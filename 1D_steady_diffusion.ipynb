## 1D diffusion equation with constant temperature boundary condition

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k=5
N=100 # Number of divisions
L=1
dx=L/N
a=k/dx
```

```python
CF=np.zeros((N,N))
for i in range(1,N-1):
    CF[i,i-1]=a
    CF[i,i]=-2*a
    CF[i,i+1]=a

CF[0,0]=-3*a 
#CF[0,0]=-a
CF[0,1]=a
CF[N-1,N-2]=a
CF[N-1,N-1]=-3*a
# CF[N-1,N-1]=-a
```

```python
T0=500  # LEFT BOUNDARY TEMPERATURE in KELVIN
TL=200  # RIGHT BOUNDARY TEMPERATURE in KELVIN
#q0=0   # LEFT BOUNDARY FLUX
#q1=100 # RIGHT BOUNDARY FLUX
BC=np.zeros(N)
BC[0]=-2*T0*a
#BC[0]=-q0
BC[N-1]=-2*TL*a
```

```python
TF=np.dot(np.linalg.inv(CF),BC)
```

```python
XC=np.arange(0,L,dx)
plt.figure(figsize=(8,6))
plt.title('Temperature distribution in a 1D domain')
plt.xlabel('X',fontsize=12)
plt.ylabel('Temperature',fontsize=12)
plt.plot(XC,TF)
plt.show()
```
